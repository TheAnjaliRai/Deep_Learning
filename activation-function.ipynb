# Explaining the non-linearily

```python
import numpy as np
import matplotlib.pyplot as plt

w=[1,2] #weights
b = 3

# creating data
x = np.linspace(-5,5,100)
y = w[0]*x + w[1]*x + b

# ploting the linear combination
plt.plot(x,y,label = "linear combination")
plt.legend
plt.show
```

```python
# ploting the sigmoid activation
sigmoid = lambda x: 1/(1+np.exp(-x))
y_sigmoid = sigmoid(y)
plt.plot(x,y_sigmoid , label = 'Sigmoid Activation')
plt.legend()
plt.show()
```

```python
# ploting the ReLU activation
relu = lambda x: np.maximum(x,0)
y_relu = relu(y)
plt.plot(x,y_relu , label = 'ReLU Activation')
plt.legend()
plt.show()
```

# Different available activation functions their comparion:

```python
import numpy as np
import matplotlib.pyplot as plt

# Activation Functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def leaky_relu_derivative(x, alpha=0.01):
    return np.where(x > 0, 1, alpha)

def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

def softmax(x):
    exp_x = np.exp(x - np.max(x))  # Subtract max for numerical stability
    return exp_x / np.sum(exp_x, axis=0)

def softmax_derivative(x):
    s = softmax(x).reshape(-1, 1)
    return np.diagflat(s) - np.dot(s, s.T)

def hard_sigmoid(x):
    return np.clip((x + 2.5) / 5, 0, 1)

def hard_sigmoid_derivative(x):
    return np.where((x > -2.5) & (x < 2.5), 0.2, 0)

# Plotting Activation Functions and their Derivatives
def plot_activation_functions(x):
    plt.figure(figsize=(12, 18))
    
    # Sigmoid and its derivative
    plt.subplot(5, 3, 1)
    plt.plot(x, sigmoid(x), label="Sigmoid")
    plt.title("Sigmoid")
    plt.grid(True)
    
    plt.subplot(5, 3, 4)
    plt.plot(x, sigmoid_derivative(x), label="Sigmoid Derivative")
    plt.title("Sigmoid Derivative")
    plt.grid(True)
    
    # ReLU and its derivative
    plt.subplot(5, 3, 2)
    plt.plot(x, relu(x), label="ReLU")
    plt.title("ReLU")
    plt.grid(True)
    
    plt.subplot(5, 3, 5)
    plt.plot(x, relu_derivative(x), label="ReLU Derivative")
    plt.title("ReLU Derivative")
    plt.grid(True)
    
    # Leaky ReLU and its derivative
    plt.subplot(5, 3, 3)
    plt.plot(x, leaky_relu(x), label="Leaky ReLU")
    plt.title("Leaky ReLU")
    plt.grid(True)
    
    plt.subplot(5, 3, 6)
    plt.plot(x, leaky_relu_derivative(x), label="Leaky ReLU Derivative")
    plt.title("Leaky ReLU Derivative")
    plt.grid(True)
    
    # Tanh and its derivative
    plt.subplot(5, 3, 7)
    plt.plot(x, tanh(x), label="Tanh")
    plt.title("Tanh")
    plt.grid(True)
    
    plt.subplot(5, 3, 10)
    plt.plot(x, tanh_derivative(x), label="Tanh Derivative")
    plt.title("Tanh Derivative")
    plt.grid(True)
    
    # Softmax (only function plot)
    plt.subplot(5, 3, 8)
    plt.plot(x, softmax(x), label="Softmax")
    plt.title("Softmax")
    plt.grid(True)

    # Hard Sigmoid and its derivative
    plt.subplot(5, 3, 9)
    plt.plot(x, hard_sigmoid(x), label="Hard Sigmoid")
    plt.title("Hard Sigmoid")
    plt.grid(True)

    plt.subplot(5, 3, 12)
    plt.plot(x, hard_sigmoid_derivative(x), label="Hard Sigmoid Derivative")
    plt.title("Hard Sigmoid Derivative")
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()

# Example usage
x = np.linspace(-10, 10, 1000)
plot_activation_functions(x)
```
